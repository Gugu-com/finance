```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error  
from sklearn.model_selection import train_test_split
```

```python
file_path = '/Users/admin/code/derksol/finance/raw_data/data_processed_wOutlier_unscaled.csv'

# Read the CSV file into a DataFrame
outlier_df = pd.read_csv(file_path)

# Drop the unnamed column in-place
outlier_df.drop(columns=['Unnamed: 0'], inplace=True)

outlier_df
```

```python
#define X and y features
X = outlier_df[['revenuePerShare' , 'netIncomePerShare' , 'operatingCashFlowPerShare' , 'freeCashFlowPerShare' , 'cashPerShare' , 'bookValuePerShare' , 'tangibleBookValuePerShare' , 'grahamNumber' , 'roic' , 'returnOnTangibleAssets' , 'grahamNetNet' , 'workingCapital' , 'tangibleAssetValue', 'netCurrentAssetValue' , 'roe']]  #drop symbol, timeframe, date , and close vs market -y target
y = outlier_df['close_vs_market']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# Create a linear regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)
```

```python

# Evaluate the model

mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error: {mae}")
```
